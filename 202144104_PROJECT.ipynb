```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '범죄_발생_장소별통계.csv'
data = pd.read_csv(file_path, encoding='euc-kr')

for idx, crime_type in enumerate(data['범죄대분류']):
    plt.figure(figsize=(16.5, 5))
    plt.bar(data.columns[1:], data.iloc[idx, 1:], color='skyblue')
    plt.ylabel('범죄 발생 건수')
    plt.title(f'{crime_type} - 장소별 범죄 발생 통계')
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '범죄_발생_시간대.csv'
data = pd.read_csv(file_path, encoding='euc-kr')

plt.figure(figsize=(18, 6))

for idx, crime_type in enumerate(data['범죄대분류']):
    plt.plot(data.columns[2:], data.iloc[idx, 2:], label=crime_type)

plt.xticks(fontsize=12)

plt.xlabel('시간대')
plt.ylabel('범죄 발생 건수')
plt.title('시간대별 범죄 발생 통계')
plt.legend(loc='upper right', bbox_to_anchor=(1.2, 1))

plt.tight_layout()
plt.show()
```
